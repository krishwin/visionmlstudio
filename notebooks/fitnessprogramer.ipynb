```python
import requests
from bs4 import BeautifulSoup

# Base URL
base_url = "https://www.fitnessprogramer.com/exercises/page/"

# List to store exercise links
exercise_links = []

# Loop through pages 1 to 73
for page in range(1, 82):
    url = base_url + str(page)
    print(f"Scraping page: {url}")
    
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find all <a> tags
        links = soup.find_all('a', href=True)
        for link in links:
            href = link['href']
            # Check if href contains '/exercises/'
            if '/exercise/' in href:
                exercise_links.append(href)
    else:
        print(f"Failed to retrieve page {page}. Status code: {response.status_code}")

# Print the collected exercise links
print("Collected Exercise Links:")
for link in exercise_links:
    print(link)
```

```python
for link in exercise_links:
    response = requests.get(link)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        img_tag = soup.findAll('img', {'src': True})
        for img in img_tag:
            if img and '.gif' in img['src']:
                print(f"Found GIF: {img['src']}")
                gif_url = img['src']
                response = requests.get(gif_url)
                if response.status_code == 200:
                    file_name = link.split('/')[-2] + ".gif"
                    with open(f"./data/fitnessprogrammer/gif/{file_name}", "wb") as gif_file:
                        gif_file.write(response.content)
                    print(f"Downloaded GIF: {file_name}")
                else:
                    print(f"Failed to download GIF from {gif_url}. Status code: {response.status_code}")
                break
    else:
        print(f"Failed to retrieve {link}. Status code: {response.status_code}")
```

```python
exercises = []
for link in exercise_links:
    exercises.append(link.split('/')[-2].replace("-", " "))
with open("./data/fitnessprogrammer/exercises.txt", "a") as file:
    for exercise in exercises:
        file.write(exercise + "\n")
```
