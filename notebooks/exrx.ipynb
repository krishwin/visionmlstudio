```python
import requests
from bs4 import BeautifulSoup

# Base URL
base_url = "https://exrx.net/Lists/Directory"

headers = {
    "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
    "accept-language": "en-US,en-GB;q=0.9,en;q=0.8,es;q=0.7,fr;q=0.6",
    "cache-control": "max-age=0",
    "if-modified-since": "Sat, 31 May 2025 09:27:37 GMT",
    "priority": "u=0, i",
    "sec-ch-ua": "Chromium;v=136, Google Chrome;v=136, Not.A/Brand;v=99",
    "sec-ch-ua-mobile": "?0",
    "sec-ch-ua-platform": "macOS",
    "sec-fetch-dest": "document",
    "sec-fetch-mode": "navigate",
    "sec-fetch-site": "same-origin",
    "sec-fetch-user": "?1",
    "upgrade-insecure-requests": "1"
}

# Send a GET request to the URL
response = requests.get(base_url,headers=headers)
muscles_urls = []
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    # Find all links to muscle groups
    muscle_links = soup.find_all('a', href=True)
    for link in muscle_links:
        href = link['href']
        if href.startswith('Exlist/') and href not in muscles_urls and "#"  in href:
            muscles_urls.append(href)
else:
    print(response.status_code,response.reason,response.text)
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options

#options = Options()
#options.page_load_strategy = 'none'  # or 'eager'
driver = webdriver.Chrome()
driver.implicitly_wait(2)
driver.get('https://exrx.net/Lists/Directory')
links = driver.find_elements(By.TAG_NAME, 'a')
```

```python
muscles_urls=[]
for link in links:
    href = link.get_attribute('href')
    #if href and href.startswith('https://exrx.net/Lists/Exlist/') and href not in muscles_urls and "#" in href:
    muscles_urls.append(href)
```

```python
import os
import requests

# Navigate to the page
driver.get("https://exrx.net/WeightExercises/WristFlexors/SLOneHandGripPlateLoaded")

# Locate the video element
video_element = driver.find_element(By.TAG_NAME, 'video')
video_url = video_element.get_attribute('src')

# Download the video
if video_url:
    response = requests.get(video_url, stream=True)
    if response.status_code == 200:
        # Save the video to a file
        video_path = "SLOneHandGripPlateLoaded.mp4"
        with open(video_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=1024):
                file.write(chunk)
        print(f"Video downloaded successfully: {video_path}")
    else:
        print(f"Failed to download video. Status code: {response.status_code}")
else:
    print("Video URL not found.")
```
