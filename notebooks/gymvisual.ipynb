```python
import os
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = "https://gymvisual.com/16-animated-gifs/s-1/media_type-animated_gifs/exercise_type-strength?"

# Directory to save the downloaded GIFs
output_dir = "./data/gymvisual/gif/"
os.makedirs(output_dir, exist_ok=True)

for i in range(7, 232):
    url1 = "https://gymvisual.com/16-animated-gifs/s-1/media_type-animated_gifs/exercise_type-strength?"+f"p={i}"
    print(f"Fetching page: {url1}")
# Fetch the page content
    response = requests.get(url1)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        
        # Find all GIF links
        gif_tags = soup.find_all("img", {"class": "replace-2x img-responsive"})
        gif_urls = [(tag["src"] , tag["alt"]) for tag in gif_tags if tag["src"].endswith(".gif")]
        # Download each GIF
        for idx, (gif_url,alt) in enumerate(gif_urls, start=1):
            try:
                gif_response = requests.get(gif_url, stream=True)
                if gif_response.status_code == 200:
                    alt = alt.replace(" ", "_").replace("/", "-")
                    gif_path = os.path.join(output_dir, f"{alt}.gif")
                    with open(gif_path, "wb") as gif_file:
                        for chunk in gif_response.iter_content(1024):
                            gif_file.write(chunk)
                    print(f"Downloaded: {gif_path}")
                else:
                    print(f"Failed to download: {gif_url}")
            except Exception as e:
                print(f"Error downloading {gif_url}: {e}")
    else:
        print(f"Failed to fetch the page. Status code: {response.status_code}")
```
