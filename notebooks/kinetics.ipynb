```python
import pandas as pd
import os
import shutil

# Read the Excel file into a DataFrame
file_path = "grouped_exercises_by_type.xlsx"
data = pd.read_excel(file_path)

# Display the first few rows of the DataFrame
data.head()
```

```python
data1 = pd.read_excel('new_grouped_exercises_by_type.xlsx')
```

```python
data['source'] = 'jefit'
data1['source'] = 'fitnessprogrammer'
```

```python
data1['file_name'] = data1['Exercise Name'].str.lower().str.replace(' ', '-')+ '.gif'
```

```python
with open("./data/fitnessprogrammer/download_links.txt", "r") as file:
    download_links = file.readlines()
download_links = [link.strip() for link in download_links]
for link in download_links:
    type = link.split('/')[-1].lower().split('.')[0].replace('-', ' ')
    data1.loc[data1['Exercise Name'].str.fullmatch(type), 'url'] = link
```

```python
for row in data1.itertuples():
    if pd.isna(row.url):
        continue
    else:
        file_name = row.file_name
        type = row[1].replace(' ', '_').replace('/','').lower().strip()
        
        # Create the directory if it doesn't exist
        dir_path = f"./data/kinetics/{type}"
        os.makedirs(dir_path, exist_ok=True)
        # Download the file
        file_path = os.path.join(dir_path, file_name)
        if not os.path.exists(file_path) and os.path.exists(f'./data/fitnessprogrammer/gif/{file_name}'):
            shutil.copyfile(f'./data/fitnessprogrammer/gif/{file_name}', file_path)
```

```python
# Read the file and split each line by '/'
with open('exercise_links.txt', 'r') as file:
    lines = file.readlines()
    split_lines = [line.strip().split('/') for line in lines]

# Display the result
split_lines[:5]  # Display the first 5 split lines as a preview
```

```python
for line in split_lines:
    type = line[3].replace('-', ' ')
    data.loc[data['Exercise Name'] == type, 'file_name'] = f"{line[2]}.gif"
    data.loc[data['Exercise Name'] == type, 'url'] = f"https://cdn.jefit.com/assets/img/exercises/gifs/{line[2]}.gif"
```

```python


for line in split_lines:
    matching_row = data[data['Exercise Name'] == line[3].replace('-', ' ')]
    if not matching_row.empty:
        # Extract the type from the matching row
        exercise_type = matching_row['Exercise Type'].values[0].replace(' ', '_').replace('/','').lower().strip()
        finename = f'./data/jefit/gif/{line[2]}.gif'
        # Define the folder path
        folder_path = f'./data/kinetics/{exercise_type}'

        # Check if the folder exists, if not, create it
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)

        # Copy the file to the folder
        if(os.path.exists(finename)):
            print(f"Copying {finename} to {folder_path}")
            shutil.copy(finename, folder_path)
```

```python
import random
import shutil
import os

# Define the source folder and train/test folders
source_folder = "./data/kinetics/"
train_folder = "./data/kinetics1/train/"
test_folder = "./data/kinetics1/test/"

# Create train and test directories if they don't exist
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# Iterate through each subfolder in the source folder
for subfolder in os.listdir(source_folder):
    subfolder_path = os.path.join(source_folder, subfolder)
    if os.path.isdir(subfolder_path):
        # Get all files in the subfolder
        files = os.listdir(subfolder_path)
        random.shuffle(files)
        # Create subfolders within train and test folders
        train_subfolder = os.path.join(train_folder, subfolder)
        test_subfolder = os.path.join(test_folder, subfolder)
        os.makedirs(train_subfolder, exist_ok=True)
        os.makedirs(test_subfolder, exist_ok=True)
        # Split files into train and test sets (80% train, 20% test)
        split_index = int(len(files) * 0.8)
        train_files = files[:split_index]
        test_files = files[split_index:]
        
        # Move files to train and test folders
        for file in train_files:
            shutil.copy(os.path.join(subfolder_path, file), os.path.join(train_subfolder, file))
        for file in test_files:
            shutil.copy(os.path.join(subfolder_path, file), os.path.join(test_subfolder, file))
```

```python
import random
import shutil
import os

# Define the source folder and train/test folders
source_folder = "./data/kinetics/"
train_folder = "./data/kinetics2/train/"
test_folder = "./data/kinetics2/test/"

# Create train and test directories if they don't exist
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# Iterate through each subfolder in the source folder
for subfolder in [ dirs for dirs, count in file_count_by_dir.items() if count > 99]:
    subfolder_path = os.path.join(source_folder, subfolder)
    if os.path.isdir(subfolder_path):
        # Get all files in the subfolder
        files = os.listdir(subfolder_path)
        random.shuffle(files)
        # Create subfolders within train and test folders
        train_subfolder = os.path.join(train_folder, subfolder)
        test_subfolder = os.path.join(test_folder, subfolder)
        os.makedirs(train_subfolder, exist_ok=True)
        os.makedirs(test_subfolder, exist_ok=True)
        # Split files into train and test sets (80% train, 20% test)
        split_index = int(len(files) * 0.8)
        train_files = files[:split_index]
        test_files = files[split_index:]
        
        # Move files to train and test folders
        for file in train_files:
            shutil.copy(os.path.join(subfolder_path, file), os.path.join(train_subfolder, file))
        for file in test_files:
            shutil.copy(os.path.join(subfolder_path, file), os.path.join(test_subfolder, file))
```

```python
import imageio

train_dir = "./data/kinetics2/"
f_count=[]
# Walk through the train directory
for root, dirs, files in os.walk(train_dir):
    if(len(files) == 0):
        print(f"Empty directory: {root}")
    for file in files:
        try:
            # Load the GIF frames using the custom loader
            gif_frames =imageio.mimread(os.path.join(root, file), memtest=False)
            #print(root, file, len(gif_frames))
            f_count.append(len(gif_frames))
        except Exception as e:
            print(f"Error loading {file} in {root}: {e}")
            continue
```

```python
import matplotlib.pyplot as plt

# Plot a histogram of frame counts
plt.hist(f_count, bins=20, edgecolor='black')
plt.title('Distribution of Frame Counts')
plt.xlabel('Frame Count')
plt.ylabel('Frequency')
plt.xlim(0, 150)
plt.show()
```
