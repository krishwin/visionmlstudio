```python
import requests
from bs4 import BeautifulSoup

# Base URL
base_url = "https://www.hevyapp.com/exercises/"

# Send a GET request to the URL
response = requests.get(base_url)
parts_urls = []
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    sects = soup.findAll('section', class_='elementor-inner-section')
    for sec in sects:
        # Find all exercise links
        links = sec.find_all('a', href=True)
        for link in links:
            if 'href' in link.attrs and 'muscle' in link['href']:
                # Append the href to parts_urls
                parts_urls.append(link['href'])
    else:
        print("No section found.")
```

```python
exercise_links = []
for url in parts_urls:
    # Send a GET request to each exercise URL
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        # Find the exercise name
        articles = soup.findAll('article', class_='elementor-post')
        for article in articles:
            links = article.findAll('a', href=True)
            for link in links:
                if 'href' in link.attrs and 'https://www.hevyapp.com/exercises/' in link['href']:
                    # Print the exercise name and URL
                    exercise_links.append(link['href'])
        nav = soup.find('nav', class_='elementor-pagination')
        if nav:
            page = nav.findAll('a', href=True)
            for p in page:
                if 'href' in p.attrs and '/page/' in p['href']:
                    # Print the exercise name and URL
                    response1 = requests.get(p['href'])
                    if response1.status_code == 200:
                        soup1 = BeautifulSoup(response1.text, 'html.parser')
                        articles1 = soup1.findAll('article', class_='elementor-post')
                        for article1 in articles1:
                            links1 = article1.findAll('a', href=True)
                            for link1 in links1:
                                if 'href' in link1.attrs and 'https://www.hevyapp.com/exercises/' in link1['href']:
                                    # Append the exercise URL to exercise_links
                                    exercise_links.append(link1['href'])
    else:
        print(f"Failed to retrieve {url}, status code: {response.status_code}")
```

```python
video_links = []
for link in exercise_links:
    # Send a GET request to each exercise link
    response = requests.get(link)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the exercise page
        soup = BeautifulSoup(response.text, 'html.parser')
        
        video = soup.find('video')
        if video:
            # Extract the video source URL
            
                video_links.append(video['src'])
            
        else:
            print(f"No video found for {link}.")
    else:
        print(f"Failed to retrieve exercise page: {link}. Status code: {response.status_code}")
```

```python
import os
import requests

# Define the target folder
target_folder = './data/heavyapp/mp4'

# Create the folder if it doesn't exist
os.makedirs(target_folder, exist_ok=True)

# Download each video link
for video_url in video_links:
    video_url = video_url.strip()  # Remove any leading/trailing whitespace
    file_name = os.path.join(target_folder, os.path.basename(video_url))
    
    # Download the video file
    response = requests.get(video_url, stream=True)
    if response.status_code == 200:
        with open(file_name, 'wb') as f:
            for chunk in response.iter_content(chunk_size=1024):
                f.write(chunk)
        print(f"Downloaded: {file_name}")
    else:
        print(f"Failed to download: {video_url}, status code: {response.status_code}")
```
