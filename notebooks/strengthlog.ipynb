```python
import requests
from bs4 import BeautifulSoup

# Base URL
base_url = "https://www.strengthlog.com/exercise-directory/"

# Send a GET request to the URL
response = requests.get(base_url)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    ol = soup.find_all('ol', class_='wp-block-list')
    print(ol[0].text)
```

```python
exercise_links = []
for li in ol:
    for a in li.find_all('a'):
        href = a.get('href')
        if href :
            exercise_links.append(href)
```

```python
with open("./data/strengthlog/exercise_links.txt", "w") as file:
    for link in exercise_links:
        file.write(link + "\n")
print(f"Found {len(exercise_links)} exercise links.")
```

```python
for link in exercise_links:
    response = requests.get(link)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        img_tag = soup.findAll('img', {'src': True})
        for img in img_tag:
            if img and '.gif' in img['src']:
                print(f"Found GIF: {img['src']}")
                gif_url = img['src']
                response = requests.get(gif_url)
                try:
                    if response.status_code == 200:
                        file_name = link.split('/')[-2] + ".gif"
                        with open(f"./data/strengthlog/gif/{file_name}", "wb") as gif_file:
                            gif_file.write(response.content)
                        print(f"Downloaded GIF: {file_name}")
                    else:
                        print(f"Failed to download GIF from {gif_url}. Status code: {response.status_code}")
                    break
                except Exception as e:
                    print(f"Error downloading GIF from {gif_url}: {e}")
                    break
    else:
        print(f"Failed to retrieve {link}. Status code: {response.status_code}")
```

```python
exercises = []
for link in exercise_links:
    exercises.append(link.split('/')[-2].replace("-", " "))
with open("./data/strengthlog/exercises.txt", "a") as file:
    for exercise in exercises:
        file.write(exercise + "\n")
```
