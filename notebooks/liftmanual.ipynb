```python
import requests
from bs4 import BeautifulSoup

# URL to scrape
url = "https://liftmanual.com/exercises/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Extract all <a> tags
a_tags = soup.find_all('a', href=True, class_='rank-math-html-sitemap__link')

# Print the extracted <a> tags
for tag in a_tags:
    print(tag)
```

```python
import os

# Create the directory if it doesn't exist
os.makedirs('./data/liftmanual', exist_ok=True)

# Write the links to the file
with open('./liftmanual/exercise_links.txt', 'w') as file:
    for tag in a_tags:
        href = tag['href']
        file.write(href + '\n')
```

```python
for tag in a_tags:
    # Extract the href attribute
    href = tag['href']
    
    # Send a GET request to the URL
    response = requests.get(href)
    name = href.split('/')[-2] + '.webp'
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract all <img> tags with .webp extension in the 'src' attribute
    img_tags = soup.find_all('img', src=lambda x: x and x.endswith('.webp'))
    
    # Print the extracted <img> tags
    for img_tag in img_tags:
        # Extract the 'src' attribute of the image tag
        img_url = img_tag['src']
        
        # Send a GET request to download the image
        img_response = requests.get(img_url)
        print(f'Downloading {img_url} to ./data/liftmanual/gifs/{name}')
        with open(f'./data/liftmanual/gifs/{name}', 'wb') as img_file:
            img_file.write(img_response.content)
```

```python
gif_tags=[]
for tag in a_tags:
    # Extract the href attribute
    href = tag['href']
    name = href.split('/')[-2] 
    if not os.path.exists(f'./data/liftmanual/gifs/{name}.webp') and not os.path.exists(f'./data/liftmanual/gifs/{name}.gif'):
        print(f'File {href} does not exist in ./data/liftmanual/gifs')
        gif_tags.append(href)
```

```python
for tag in gif_tags:
    # Extract the href attribute
    href = tag
    name = href.split('/')[-2] + '.gif'

    if not os.path.exists(f'./data/liftmanual/gifs/{name}'):
        # Send a GET request to the URL
        response = requests.get(href)
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract all <img> tags with .webp extension in the 'src' attribute
        img_tags = soup.find_all('img', src=lambda x: x and x.endswith('.gif'))
        
        # Print the extracted <img> tags
        for img_tag in img_tags:
            # Extract the 'src' attribute of the image tag
            img_url = img_tag['src']
            
            # Send a GET request to download the image
            img_response = requests.get(img_url)
            print(f'Downloading {img_url} to ./data/liftmanual/gifs/{name}')
            with open(f'./data/liftmanual/gifs/{name}', 'wb') as img_file:
                img_file.write(img_response.content)
```
